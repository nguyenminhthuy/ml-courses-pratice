# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Set up matplot
fig, ax = plt.subplots(figsize=(5,5))

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. save and show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

# 2 options for subplots

```python
# subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))

# plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_price.keys(), nut_butter_price.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                        ncols=2,
                        figsize=(10,5))

# plot to each different axis
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_price.keys(), nut_butter_price.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use? pyplot vs matplotlib OO method

* when plotting something quickly, okay to use the pyplot method

* when plotting something more avanced, use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] >50]
over_50
```

```python
over_50.head()
```

```python
#pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(10,6));
```

```python
# OO method
# ax: hey, we're using the old method and we want you to plot this data
fix, ax =plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

```python
fix, ax =plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
ax.set_xlim([45,100]);
```

```python
over_50.target.values
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

# customize the plot
ax.set(title="Heart disease and Cholesterol levels",
      xlabel="Age",
      ylabel="Cholesterol");

# add a legend
# legend_elements() get from c=over_50["target"]
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10))

# add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

# customize ax0
ax0.set(title="heart disease and Cholesterol levels",
       xlabel="Age",
       ylabel="Cholesterol")

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

#------------------------------------
# add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"])

# customize ax0
ax1.set(title="Heart disease and Max heart rate",
       xlabel="Age",
       ylabel="Max heart rate");

# add a legend to ax0
ax1.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# add a title to the figure
fig.suptitle("Heart disease analysis", fontsize=16, fontweight="bold");
```

```python
# set the stype
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="plasma"); 
                    # this changes the color scheme (winter, summer, etc)

# customize the plot
ax.set(title="Heart disease and Cholesterol levels",
      xlabel="Age",
      ylabel="Cholesterol");

# add a legend
# legend_elements() get from c=over_50["target"]
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
# set the stype
plt.style.use('seaborn-whitegrid')

# subplot of chol, age, thalach
fig, (ax0, ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                            sharex=True)

# add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter")

# customize ax0
ax0.set(title="heart disease and Cholesterol levels",
       xlabel="Age",
       ylabel="Cholesterol")

# change the x axis limits
ax0.set_xlim([50, 80])

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

#------------------------------------
# add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap="summer")

# customize ax0
ax1.set(title="Heart disease and Max heart rate",
       xlabel="Age",
       ylabel="Max heart rate");

# change the x axis limits
ax1.set_xlim([50, 80])

# add a legend to ax0
ax1.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# add a title to the figure
fig.suptitle("Heart disease analysis", fontsize=16, fontweight="bold");
```
